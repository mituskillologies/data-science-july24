```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([12, 23, 34, 39, 45, 48, 53, 60, 69, 76])
b = np.array([20, 30, 35, 38, 46, 51, 55, 59, 65, 70])
c = np.array([76, 70, 67, 64, 58, 55, 47, 41, 38, 25])
```

#### Line Plot

```python
plt.plot(a, b);
```

```python
plt.plot(a, c);
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(a, b, marker = 'o', mfc = 'red', ms = 7,
        ls = '--');
plt.grid()

# marker = >, <, ^, H, *, o
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(a, c, marker = 'o', mfc = 'red', ms = 7,
        ls = '--');
plt.grid()
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(a, b, marker = 'o', mfc = 'red', ms = 7,
        ls = '--', label = '2023');
plt.plot(a, c, marker = 'o', mfc = 'yellow', ms = 7,
        ls = '--', color = 'black', label = 2024);
plt.grid()
plt.legend();
```

#### subplot

```python
plt.figure(figsize=(16, 9))

plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(a, b, marker = 'o', mfc = 'red', ms = 7,
        ls = '--');
plt.grid()

plt.subplot(2,2,2)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(a, c, marker = 'o', mfc = 'red', ms = 7,
        ls = '--');
plt.grid()
```

#### Bar Graph

```python
courses = ['java', 'c', 'python', 'dart', 'kotlin']
students = [56, 70, 63, 68, 52]
```

```python
plt.title('bar graph')
plt.xlabel('course name')
plt.ylabel('no. of students')
plt.bar(courses, students, color = 'black', width = 0.3);
```

### scatter plot

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.scatter(a, b, marker = '*');
plt.grid()
```

#### Pie chart

```python
plt.title('pie chart')

mycol = ['cyan', 'orange', 'blue', 'green', 'skyblue']
plt.pie(students, labels = courses, autopct = '%.2f%%', colors = mycol,
       shadow = True, labeldistance= 1.1);
```

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(a, b, marker = 'o', mfc = 'red', ms = 7,
        ls = '--', label = '2023');
plt.plot(a, c, marker = 'o', mfc = 'yellow', ms = 7,
        ls = '--', color = 'black', label = 2024);
plt.grid()
plt.legend();

plt.subplot(2,2,2)
plt.title('bar graph')
plt.xlabel('course name')
plt.ylabel('no. of students')
plt.bar(courses, students, color = 'black', width = 0.3);

plt.subplot(2,2,3)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.scatter(a, b, marker = '*');
plt.grid()

plt.subplot(2,2,4)
plt.title('pie chart')
mycol = ['cyan', 'orange', 'blue', 'green', 'skyblue']
plt.pie(students, labels = courses, autopct = '%.2f%%', colors = mycol,
       shadow = True, labeldistance= 1.1);

plt.savefig('myplot.png')
```

#### Box Plot

```python
grades = [67, 70, 71, 59, 63, 73, 64, 97]
```

```python
plt.boxplot(grades);
plt.grid()
```

```python
q3 = np.percentile(grades, 75)
q1 = np.percentile(grades, 25)
```

```python
q3, q1
```

```python
iqr = q3 - q1
```

```python
iqr
```

```python
upper_bound = q3 + (1.5 * iqr)
```

```python
lower_bound = q1 - (1.5 * iqr)
```

```python
upper_bound, lower_bound
```
